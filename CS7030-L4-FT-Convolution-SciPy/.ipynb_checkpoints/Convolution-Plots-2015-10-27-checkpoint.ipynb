```python
'''
Convolve your 2-D function from the previous lab assignment with the cat image image. 
Both in Spatial frequency and spatial domains.

Make some plots demonstrating differences between Spatial Frequency and Spatial domain convolution for various
sizes of image and kernel.

'''
%matplotlib notebook

import numpy as np

from scipy import misc
from scipy import ndimage
from scipy import signal

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg

#Import Cat
img = mpimg.imread('StripedCat.png')

#Rectangle
start_time = -2
stop_time = 2
time_to_plot = 4 # seconds
sample_rate = 25 # In Hz, the number of samples per second
total_number_of_sample = sample_rate * time_to_plot #+ 1 #Odd numbers get you zero with linspace
t = np.linspace(start_time, stop_time, total_number_of_sample) #Odd numbers get you zero with linspace
rect = np.linspace(0,1,total_number_of_sample)
for i in range(len(t)):
    if((t[i] > -0.5) & (t[i] < 0.5)):
        rect[i] = 1
    elif((t[i] == 0.5) &(t[i] == -0.5)):
        rect[i] = 0.5
    else:
        rect[i] = 0

#setup_graph(x_label='time (in seconds)', y_label='amplitude', title='Rectangle in time domain')
#plt.axis([-2.1, 2.1, 0,1.1])
#plt.plot(t, rect)
#plt.show()
rect1 = np.asmatrix(rect)
rect2D=np.dot(np.transpose(rect1),rect1)
plt.matshow(rect2D)
plt.show()

#Convolve with image
rectcon= signal.convolve2d(img, rect2D)

rectconP = plt.figure()
axirectconP = rectconP.add_subplot(1,1,1)
axirectconP.imshow(np.abs(rectcon), cmap=cm.Greys_r)

```

```python
# Signum:
start_time = -2
stop_time = 2
time_to_plot = 4 # seconds
sample_rate = 25 # In Hz, the number of samples per second
total_number_of_sample = sample_rate * time_to_plot + 1 #Odd numbers get you zero with linspace
t = np.linspace(start_time, stop_time, total_number_of_sample) #Odd numbers get you zero with linspace
signum = np.copysign(1,t) # Define coysign is closest function in python to Signum array
signum[int((sample_rate * time_to_plot)/2)] = 0 #add zero because copysign turns zero to 1

signum1 = np.asmatrix(signum)
signum2D=np.dot(np.transpose(signum1),signum1)

plt.matshow(signum2D)
plt.show()

#Convolve with image
signumcon= signal.convolve2d(img, signum2D)

signumconP = plt.figure()
axisignumconP = signumconP.add_subplot(1,1,1)
axisignumconP.imshow(np.abs(signumcon), cmap=cm.Greys_r)
```

```python
# Triangle:
start_time = -2
stop_time = 2
time_to_plot = 4 # seconds
sample_rate = 25 # In Hz, the number of samples per second
total_number_of_sample = sample_rate * time_to_plot #+ 1 #Odd numbers get you zero with linspace
t = np.linspace(start_time, stop_time, total_number_of_sample) #Odd numbers get you zero with linspace
triangle= np.linspace(0,1,total_number_of_sample)
for i in range(len(t)):
    if((t[i] > -1.0) & (t[i] < 1.0)):
        triangle[i] = 1 - abs(t[i])
    else:
        triangle[i] = 0

triangle1 = np.asmatrix(triangle)
triangle2D=np.dot(np.transpose(triangle1),triangle1)

plt.matshow(triangle2D, cmap=cm.Greys_r)
plt.show()

#Convolve with image
trianglecon= signal.convolve2d(img, triangle2D)

triangleconP = plt.figure()
axitriangleconP = triangleconP.add_subplot(1,1,1)
axitriangleconP.imshow(np.abs(trianglecon), cmap=cm.Greys_r)
```

```python
# Impulse/Dirac delta:
start_time = -2
stop_time = 2
time_to_plot = 4 # seconds
sample_rate = 25 # In Hz, the number of samples per second
total_number_of_sample = sample_rate * time_to_plot + 1 #Odd numbers get you zero with linspace
t = np.linspace(start_time, stop_time, total_number_of_sample) #Odd numbers get you zero with linspace
dirac_impulze = np.linspace(0,1,total_number_of_sample)

for i in range(len(t)):
    if(t[i] == 0):
        dirac_impulze[i]  = 100000
    else:
        dirac_impulze [i] = 0
        
#Generate 2D dirac

dirac_impulze1 = np.asmatrix(dirac_impulze)
dirac_impulze2D=np.dot(np.transpose(dirac_impulze1),dirac_impulze1)

plt.matshow(dirac_impulze2D)
plt.show()

        
#Convolve with image
dirac_impulzecon= signal.convolve2d(img, dirac_impulze2D)

dirac_impulzeconP = plt.figure()
axidirac_impulzeconP = dirac_impulzeconP.add_subplot(1,1,1)
axidirac_impulzeconP.imshow(np.abs(dirac_impulzecon), cmap=cm.Greys_r)
```

```python
# Combulusion - Dirac delta impul train:
start_time = -2
stop_time = 2
time_to_plot = 4 # seconds
sample_rate = 100 # In Hz, the number of samples per second
total_number_of_sample = sample_rate * time_to_plot + 1 #Odd numbers get you zero with linspace
t = np.linspace(start_time, stop_time, total_number_of_sample) #Odd numbers get you zero with linspace
from scipy.special import comb

dirac_comb = np.linspace(0,1,total_number_of_sample)
k = 10 # set time between impulses, everysecond

for i in range(len(t)):
    if (i % k) == 0:
        dirac_comb[i] = 1000
    else:
        dirac_comb[i] = 0

#Generate 2D Comb

dirac_comb1 = np.asmatrix(dirac_comb)
dirac_comb2D=np.dot(np.transpose(dirac_comb1),dirac_comb1)

plt.matshow(dirac_comb2D,cmap=cm.Greys_r)
plt.show()

#Convolve with image
dirac_combcon= signal.convolve2d(img, dirac_comb2D)

dirac_combconP = plt.figure()
axidirac_combconP = dirac_combconP.add_subplot(1,1,1)
axidirac_combconP.imshow(np.abs(dirac_combcon), cmap=cm.Greys_r)
```
